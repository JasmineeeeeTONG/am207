## Question 1: We'll Always Have that Night Sampling in Monte Carlo

**Coding required**


Let $X$ be a random variable with distribution described by the following pdf:

$$
f_X(x) = \begin{cases}
\frac{1}{12}(x-1), &1\leq x\leq 3\\
-\frac{1}{12}(x-5), &3< x\leq 5\\
\frac{1}{6}(x-5), &5< x\leq 7\\
-\frac{1}{6}(x-9), &7< x\leq 9\\
0, &otherwise
\end{cases}
$$

Let $h$ be the following function of $X$:

$$h(X) = \frac{1}{3\sqrt{2}\pi}\mathrm{exp}\left\{ -\frac{1}{18}\left( X - 5\right)^2\right\}$$


Compute $\mathbb{E}[h(X)]$ via Monte Carlo simulation using the following sampling methods:

**1.1.** Inverse Transform Sampling

**1.2.** Rejection Sampling with a uniform proposal distribution (rejection sampling in a rectangular box with uniform probability of sampling any x)


### Answer 1.1, 1.2

CDF of X, $F_X(x)$:
$$
F_X(x) = \begin{cases}
0, &x < 1\\
\frac{1}{24} x^2 - \frac{1}{12}x + \frac{1}{24}, &1\leq x\leq 3\\
-\frac{1}{24} x^2 + \frac{5}{12}x - \frac{17}{24}, &3< x\leq 5\\
\frac{1}{12} x^2 - \frac{5}{6}x + \frac{29}{12}, &5< x\leq 7\\
-\frac{1}{12} x^2 + \frac{3}{2}x - \frac{23}{4}, &7< x\leq 9\\
1, &x > 9\\
\end{cases}
$$

Inverse CDF of X, $F^{-1}_X(y)$:
$$
F^{-1}_X(y) = \begin{cases}
\sqrt{24y} + 1, &0\leq y\leq \frac{1}{6}\\
5 - \sqrt{8 - 24y}, &\frac{1}{6}< y\leq \frac{1}{3}\\
5 + \sqrt{12y - 4}, &\frac{1}{3}< y\leq \frac{2}{3}\\
9 - \sqrt{12 - 12y}, &\frac{2}{3}< y\leq 1
\end{cases}
$$

```python
# 1.1, 1.2
def PDF_X(x):
    if x < 1: return 0
    if x >= 1 and x <= 3: return 1/12*(x-1)
    if x > 3 and x <= 5: return -1/12*(x-5)
    if x > 5 and x <= 7: return 1/6*(x-5)
    if x > 7 and x <= 9: return -1/6*(x-9)
    return 0

def CDF_X(x):
    if x <= 1: return 0
    if x > 1 and x <= 3: return 1/24*x**2 - 1/12*x + 1/24
    if x > 3 and x <= 5: return -1/24*x**2 + 5/12*x - 17/24
    if x > 5 and x <= 7: return 1/12*x**2 - 5/6*x + 29/12
    if x > 7 and x <= 9: return -1/12*x**2 + 3/2*x - 23/4
    return 1

def INV_CDF_X(y):
    if y < 0 or y > 1: raise Exception('y must be within (0, 1).')
    if y <= 1/6: return np.sqrt(24*y) + 1
    if y > 1/6 and y <= 1/3: return 5 - np.sqrt(8 - 24*y)
    if y > 1/3 and y <= 2/3: return 5 + np.sqrt(12*y - 4)
    if y > 2/3 and y <= 1: return 9 - np.sqrt(12 - 12*y)
    return 1

x_min = 0
x_max = 10
N = 10000
x_lin = np.linspace(x_min, x_max, N)
PDF_x_lin = np.array([PDF_X(xi) for xi in x_lin])
CDF_x_lin = np.array([CDF_X(xi) for xi in x_lin])

y_lin = np.linspace(0, 1, N)
INV_CDF_lin = np.array([INV_CDF_X(yi) for yi in y_lin])

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].plot(x_lin, PDF_x_lin)
axes[1].plot(x_lin, CDF_x_lin)
axes[2].plot(x_lin, INV_CDF_lin)

axes[0].set_title('PDF', fontsize=16)
axes[1].set_title('CDF', fontsize=16)
axes[2].set_title('INV_CDF', fontsize=16)

axes[0].set_xlabel('x', fontsize=14)
axes[0].set_ylabel('y', fontsize=14)
axes[1].set_xlabel('x', fontsize=14)
axes[1].set_ylabel('y', fontsize=14)
axes[2].set_xlabel('y', fontsize=14)
plt.tight_layout()
```

```python
# Inverse Transform sampling
CDF_x_min = CDF_X(x_min)
CDF_x_max = CDF_X(x_max)

R_unif = np.random.uniform(CDF_x_min, CDF_x_max, N)
X_inv_samp = np.array([INV_CDF_X(ru) for ru in R_unif]) 

# Rejection Sampling
y_max = 0.5
accepted = 0 # the number of accepted samples
X_rej_samp = np.zeros(N)
count = 0 # the total count of proposals
while (accepted < N):
    x = np.random.uniform(x_min, x_max) # x ~ uniform[xmin, xmax)
    y = np.random.uniform(0, y_max) # y ~ uniform[0, ymax)
    if y < PDF_X(x): # rejection testing
        X_rej_samp[accepted] = x
        accepted += 1
    count +=1

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
pd.Series(X_inv_samp).plot(kind='hist', bins=25, density=True, alpha=0.5, ax=axes[0],
                       title='Inverse Transform Samples Density')
pd.Series(X_rej_samp).plot(kind='hist', bins=25, density=True, alpha=0.5, ax=axes[1],
                       title='Rejection Samples Density')
plt.tight_layout()
```

```python
def h_X(x):
    return 1/(3*np.sqrt(2)*np.pi) * np.exp(-1/18*(x-5)**2)

E_h_inv_samp = np.array([h_X(xi) for xi in X_inv_samp]).mean()
E_h_rej_samp = np.array([h_X(xi) for xi in X_rej_samp]).mean()

print('E[h(x)] - Inverse Transform Samples: ', E_h_inv_samp)
print('E[h(x)] - Rejection Samples: ', E_h_rej_samp)
```

```python
df_chall = pd.read_csv('chall.txt', sep='   ', header=None, names=['t', 'Y'])
df_chall.head()
```

```python
ax = sns.regplot(df_chall['t'], df_chall['Y'])
ax.set_title('Explorative Regression Plot: Failure VS. Temperature')
plt.tight_layout()
```

```python
def sigmoid_theta(theta_1, theta_2, t):
    z = theta_1 + theta_2 * t
    return 1/(1+ np.exp(-z))

def prior_x_likelihood(theta_1, theta_2, X, Y, proposal='uniform'):
    if proposal == 'uniform':
        prior = (1/50) * (1)
    elif proposal == 'normal':
        prior = norm.pdf(theta_1, loc=15, scale=5) * norm.pdf(theta_2, loc=-1, scale=1)
    Z = theta_1 + theta_2 * X
    P = 1 / (1+ np.exp(-Z))
    likelihood = (P**Y * (1-P)**(1-Y)).prod()
    return prior * likelihood

def get_max_post(proposal):
    n_samp = 10
    theta_1_unif = np.linspace(-25, 45, n_samp)
    theta_2_unif = np.linspace(-4, 2, n_samp)
    M = 0 
    for _i in range(n_samp):
        theta_1_i = theta_1_unif[_i]
        for _j in range(n_samp):
            v = prior_x_likelihood(theta_1_i, theta_2_unif[_j], df_chall['t'], df_chall['Y'], proposal)
            M = max(v, M)
    return M

def rej_sampling(X, Y, post_max, N, proposal='uniform'):
    accepted = 0
    count = 0
    intercept_samples, coef_samples = [],[]
    
    while (accepted < N):
        t1 = np.random.uniform(-25, 45)
        t2 = np.random.uniform(-4, 2)
        y = np.random.uniform(0, 1)
        if y < prior_x_likelihood(t1, t2, X, Y, proposal) / post_max:
            intercept_samples.append(t1)
            coef_samples.append(t2)
            accepted += 1
        count += 1
        if count % 100 == 0:
            print('accepted = {}, count = {}'.format(accepted, count), end='\r')
    
    return intercept_samples, coef_samples
```

```python
M_norm = get_max_post('normal')
M_norm
```

```python
t1_samples_norm, t2_samples_norm = rej_sampling(df_chall['t'], df_chall['Y'], M_norm, 5000, proposal='normal')
```

```python
# 2.3
ones_col = np.ones((df_chall.shape[0], 1))
t_ones = np.concatenate((ones_col, df_chall[['t']]), axis=1)
logit_results = sm.Logit(df_chall['Y'], t_ones).fit()
result_CI = logit_results.conf_int(alpha=0.32)
```

```python
result_CI.rename(index={'const':'intercept', 'x1':'coef'}, columns={0: 'sm.logit.32%', 1: 'sm.logit.68%'}, 
                 inplace=True)
result_CI
```

### Answer 3.1

The CDF of Y:
$$F_Y(y\ \vert\ n, \theta) = \int_{-\infty}^y f_Y(y\ |\ n, \theta) dy = P(Y \leq y) = P(\text{max}(X_{1:n}) \leq y)
= \left\{
    \begin{array}{ll}
        0 & y < 0\\
        {(\frac{y}{\theta})}^n & 0 \leq y \leq \theta\\
        1 & y > \theta
    \end{array}
   \right.
$$

The PDF of Y:
$$\Rightarrow
f_Y(y\ |\ n, \theta) = \frac{\partial F_Y(y\ \vert\ n, \theta)}{\partial y}
= \left\{
    \begin{array}{ll}
        0 & y < 0\\
        \frac{n y^{n-1}}{\theta^n} & 0 \leq y \leq \theta\\
        0 & y > \theta
    \end{array}
   \right.
$$

---
### Answer 3.2
$Z \equiv n \cdot (\theta - Y) \Rightarrow Y = \theta - \frac{Z}{n}$
$$
\begin{align} 
F_Z(z\ \vert\ n, \theta) = P(Z \leq z)
&= P(Y \geq \theta - \frac{z}{n})\\
&= P(Y > \theta - \frac{z}{n}) \text{ (as }F_Y\text{ is continuous)}\\
&= 1 - P(Y \leq \theta - \frac{z}{n})\\
&= 1 - F_Y(y\ \vert\ n, \theta)
\end{align}
$$

The CDF of Z:
$$F_Z(z\ \vert\ n, \theta) 
= \left\{
    \begin{array}{ll}
        0 & z < 0\\
        1 - (1-\frac{z}{n\theta})^n & 0 \leq z \leq n\theta\\
        1 & z > n\theta
    \end{array}
   \right.
$$

The PDF of Z:
$$f_Z(z\ \vert\ n, \theta) 
= \left\{
    \begin{array}{ll}
        0 & z < 0\\
        \frac{1}{\theta}(1-\frac{z}{n\theta})^{n-1} & 0 \leq z \leq n\theta\\
        0 & z > n\theta
    \end{array}
   \right.
$$

---
### Answer 3.3
$$
\lim_{n \rightarrow \infty} \Big[ \frac{1}{\theta}(1-\frac{z}{n\theta})^{n-1} \Big]
= \lim_{n \rightarrow \infty} \Big[ \frac{1}{\theta}(1-\frac{z}{n\theta})^{-1}(1-\frac{z}{n\theta})^n \Big]
= \frac{1}{\theta} e^{-\frac{z}{\theta}}
$$

$$\lim_{n \rightarrow \infty} \Big[ f_Z(z\ \vert\ n, \theta) \Big]
= \left\{
    \begin{array}{ll}
        0 & z < 0\\
        \frac{1}{\theta} e^{-\frac{z}{\theta}} & 0 \leq z \leq n\theta
    \end{array}
   \right.
$$

As $n \rightarrow \infty$, 
$f_Z(z\ \vert\ n, \theta) \rightarrow$ an exponential distribution with mean $\theta$.

```python
def PDF_Z_lim(z, theta):
    if z<0: return 0
    return 1/theta * np.exp(-z/theta)

theta_q3 = 1000
z_lin = np.linspace(0, 10000, 1000)
PDF_z_lin = np.array([PDF_Z_lim(zi, theta_q3) for zi in z_lin])

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(z_lin, PDF_z_lin)
ax.set_title(r'Limiting PDF of Z, $\theta$={}'.format(theta_q3))
plt.tight_layout()
```

### Answer 3.9

$$
\begin{align}
p(\theta |\ y, n) \propto p(\theta) (y\ |\ n, \theta)
&= \frac{\alpha \beta^\alpha}{\theta^{\alpha+1}} \mathcal{I}_{\{\theta \geq \beta\}} \Big( \frac{n y^{n-1}}{\theta^n} \Big) \mathcal{I}_{\{\theta \geq y\}}\\
&= \frac{\alpha \beta^\alpha}{\theta^{\alpha+1}} \frac{n}{\theta^n} (y^{n-1}) \mathcal{I}_{\{\theta \geq max(y,\beta)\}}\\
&\propto \frac{\mathcal{I}_{\{\theta \geq max(y,\beta)\}}}{\theta^{\alpha+1+n}}\\
&\propto \text{Pareto(}n+\alpha, \max (\beta, Y)\text{)}
\end{align}
$$

### Answer 3.10

```python
# 3.10
def PDF_Pareto(theta, alpha, beta):
    if theta < beta: return 0
    return alpha * (beta/theta) ** alpha / theta

def CDF_Pareto(theta, alpha, beta):
    if theta < beta: return 0
    return 1 - (beta/theta)**alpha

def INV_CDF_Pareto(v, alpha, beta):
    return beta / ((1-v)**(1/alpha))

alpha = 2
beta = 1

theta_min = 0
theta_max = 5
N_samples = 1000
theta_lin = np.linspace(theta_min, theta_max, N_samples)
PDF_Pareto_lin = np.array([PDF_Pareto(ti, alpha, beta) for ti in theta_lin])
CDF_Pareto_lin = np.array([CDF_Pareto(ti, alpha, beta) for ti in theta_lin])

v_lin = np.linspace(1e-4, 1-1e-4, N_samples)
INV_CDF_Pareto_lin = np.array([INV_CDF_Pareto(vi, alpha, beta) for vi in v_lin])

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(theta_lin, PDF_Pareto_lin)
axes[1].plot(theta_lin, CDF_Pareto_lin)
axes[2].plot(theta_lin, INV_CDF_Pareto_lin)

axes[0].set_title('Pareto PDF', fontsize=16)
axes[1].set_title('Pareto CDF', fontsize=16)
axes[2].set_title('Pareto INV_CDF', fontsize=16)

axes[0].set_xlabel(r'$\theta$', fontsize=14)
axes[1].set_xlabel(r'$\theta$', fontsize=14)
axes[2].set_xlabel(r'v', fontsize=14)
plt.tight_layout()
```

### Answer 3.11

**How does the 68% HPD compare with the confidence interval generated from bootstrapping? Why doesn't the bayesian interval construction suffer the same concerns you noted in 3.6**

The 68% Highest Posterior Density came from the 68% left-most part of the right-skewed distribution, while the boostrapping confidence interval contains the middle 68% samples. Since the Bayesian posterior of $\theta$, 

$$\text{Pareto(}n+\alpha, \max (\beta, Y)\text{)}$$

has a domain starting from $Y = \hat \theta$, the true $\theta$ is guaranteed to be captureded within this Highest Posterior Region.
