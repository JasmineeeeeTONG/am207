## Question 1: Metropoflix and Chill (What's your Net Worth)?

**coding required**

Suppose we ask you to memorize the order of the top five movies on IMDB. When we quiz you on the order afterwards, you may not recall the correct order, but the mistakes you make in your recall can be modeled by simple probabilistic models.
  
Let's say that the top five movies are:  
1. *The Shawshank Redemption*
2. *The Godfather*
3. *The Godfather: Part II*
4. *Black Panther*
5. *Pulp Fiction*

Let's represent this ordering by the vector $\omega = (1,2,3,4,5)$. 

If you were to mistakenly recall the top five movies as:
2. *The Godfather*
3. *The Godfather: Part II*
5. *Pulp Fiction*
4. *Black Panther*
1. *The Shawshank Redemption*

We'd represent your answer by the vector $\theta = (2,3,5,4,1)$.

Unfortunately, your answer is wrong.  Fortunately (for our purposes) we have a way of quantifying just how wrong. Define the Hamming distance between two top five rankings, $\theta, \omega$, as follows:
$$d(\theta, \omega) = \sum_{i=1}^5 \mathbb{I}_{\theta_i\neq \omega_i},$$ 
where $\mathbb{I}_{\theta_i\neq \omega_i}$ is an indicator function that returns 1 if $\theta_i\neq \omega_i$, and 0 otherwise.

For example, the Hamming distance between your answer and the correct answer is $d(\theta, \omega)=4$, because you only ranked *Black Panther* correctly. 

Finally, let's suppose that the probability of giving a particular answer (expressed as $\theta$) is modeled as
$$ p(\theta \,|\, \omega, \lambda) \propto  e^{-\lambda\, d(\theta,\, \omega)}$$
where $\lambda$ can be thought of as an inverse temperature

1.1. Implement a Metropolis sampler to produce sample guesses from 500 individuals, with the $\lambda$ values, $\lambda=0.2, 0.5, 1.0$. What are the top five possible guesses?

1.2. Compute the probability that *The Shawshank Redemption* is ranked as the top movie (ranked number 1) by the Metropolis algorithm sampler. Compare the resulting probabilities for the various $\lambda$ values. 

1.3. How does $\lambda$ affect the probability that *The Shawshank Redemption* is ranked as the top movie?

```python
# 1.1 metropolis sampling
def p(theta, w, lam):
    return np.exp(-lam * (theta!=w).sum())

def q_draw(theta): 
    return np.random.permutation(theta)

def metropolis(p, q_draw, w, lam, N, start, burnin=0.1, thin=2):
    samples = np.empty((N, 5))
    x_prev = start
    accepted = np.zeros((N,))
    
    for i in range(N):
        x_star = q_draw(x_prev)
        p_star = p(x_star, w, lam)
        p_prev = p(x_prev, w, lam)
        pdf_ratio = p_star / p_prev
        if np.random.uniform() < min(1, pdf_ratio):
            samples[i] = x_star
            x_prev = x_star
            accepted[i] = 1
        else:
            samples[i, :] = x_prev
        if i % 100 == 0:
            print('i = {}'.format(i), end='\r')
            
    # throw away burnin and thin
    samples = samples[int(burnin*N)::thin]
    accepted_count = np.sum(accepted[int(burnin*N)::thin])
            
    return samples, accepted_count/len(samples)
```

```python
# draw samples from Metropolis
w = np.array([1, 2, 3, 4, 5])
lam_arr = np.array([0.2, 0.5, 1.0])
theta_0 = np.array([2, 3, 5, 4, 1])

target_N = 500
burnin = 0.1
thin = 50
N = int((target_N * thin) / (1 - burnin))
print('target_N = {}, N = {}'.format(target_N, N))

samples_dict = {}
for lam in lam_arr:
    samples_dict[lam] = {}
    samples, acp_rate = metropolis(p, q_draw, w, lam, N, theta_0, burnin=burnin, thin=thin)
    samples_dict[lam]['samples'] = samples
    samples_dict[lam]['acp_rate'] = acp_rate
    print('lambda = {} | acp_rate = {}'.format(lam, acp_rate))
    
```

```python
print('top 5 guesses with counts')
for lam in lam_arr:
    lam_name = 'lambda = {}'.format(lam)
    df = pd.Series([np.str(s) for s in samples_dict[lam]['samples']], name=lam_name).to_frame()
    df['count'] = 1
    display(df.groupby(lam_name)['count'].sum().sort_values(ascending=False).head())
    print('--------------------')
```

```python
# 1.2 probability of (theta[0] == 1)
for lam in lam_arr:
    samples = samples_dict[lam]['samples']
    print('lambda = {} | p(guess[0] = 1) = {}'.format(lam, (samples[:, 0]==1).mean()))
```

### Answer 1.3

Larger $\lambda$ results in more punishment. Therefore, the larger $\lambda$ values, the more correct guesses and the higher probability that The Shawshank Redemption is ranked as the top movie.

```python
# read data
df_iris = pd.read_csv('iris.csv')

# add class label 0/1
df_iris['Y'] = df_iris['class'].apply(lambda x: int(x.split('-')[-1] == 'virginica'))

# add ones
df_iris['const'] = np.ones((len(df_iris),))
df_iris.head()
```

```python
# extract features
features = ['const'] + df_iris.columns.values[:4].tolist()
print('feature dimension = {}'.format(len(features)))

# 2.1 stratified train/test split
X_train, X_test, y_train, y_test = train_test_split(df_iris[features], df_iris['Y'], stratify=df_iris['Y'],
                                                    test_size=0.4, random_state=100)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

### Answer 2.2 

$$
\begin{align}
p(\beta \vert Y, X) &\propto p(\beta) p(Y, X \vert \beta) \\
&= \frac{1}{\sqrt{2\pi \vert \sigma^2 I \vert}} \exp \left\{{\beta^T (\sigma^2 I)^{-1} \beta} \right\}
\left( \prod_i^{N}p(y_i = 1)^{y_i} p(y_i = 0)^{1-y_i} \right)
\end{align}
$$


```python
X_shared = shared(X_train.values)

with Model() as iris_model:
    # Model parameter
    beta = pm.Normal('beta', mu=0, sd=10, shape=len(features))

    # Calculate probabilities of p(Y=1)
    p = pm.Deterministic('p', invlogit(tt.dot(X_shared, beta)))

    # Data likelihood
    Y = pm.Bernoulli('Y', p=p, observed=y_train.values)
```

```python
pm.model_to_graphviz(iris_model)
```

```python
# 2.3 parameters MAP estimates
map_params = pm.find_MAP(model=iris_model)
map_params['beta']
```

```python
# 2.4 sample with posteior betas
with iris_model:
    train_trace = pm.sample(5000, step=pm.NUTS(), start=map_params)
```

```python
# 2.5 visualize beta posteriors
pm.plot_posterior(train_trace, varnames=['beta'])
plt.tight_layout()
```

```python
# 2.5 visualize beta posteriors - histogram & trace
pm.traceplot(train_trace, varnames=['beta'])
plt.tight_layout()
```

```python
# 2.5 visualize beta posteriors - autocorrelations
pm.autocorrplot(train_trace, varnames=['beta'])
plt.tight_layout()
```

```python
# 2.6 beta posterior mean
beta_post_mean = train_trace.get_values('beta').mean(axis=0)
beta_post_mean
```

```python
train_trace.get_values('beta').shape
```

```python
# 2.7 Plot posterior distribution, mean & MAP at 2 data points
y_train.head()
```

```python
idx_y1 = 0 # select 1 data point with y=1
idx_y0 = 2 # select 1 data point with y=0

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
plt.suptitle('Histogram of Posterior at the 2 data points', fontsize=16, weight='heavy')
plt.subplots_adjust(top=0.8)

ax[0].hist(train_trace['p'][:, idx_y0], alpha=0.5, bins=25, label='posterior p')
ax[0].axvline(map_params['p'][idx_y0], c='r', label='MAP')
ax[0].axvline(train_trace['p'][:, idx_y0].mean(), c='b', label='posterior mean')
ax[0].set_title('Y = 0', fontsize=16)
ax[0].legend(fontsize=12)

ax[1].hist(train_trace['p'][:, idx_y1], alpha=0.5, bins=25, label='posterior p')
ax[1].axvline(map_params['p'][idx_y1], c='r', label='MAP')
ax[1].axvline(train_trace['p'][:, idx_y1].mean(), c='b', label='posterior mean')
ax[1].set_title('Y = 1', fontsize=16)
ax[1].legend(fontsize=12)

plt.show()
```

```python
# 2.8 distribution of p's from different beta estimates
y_train_PP = pm.sample_ppc(train_trace, model=iris_model, samples=10000)

p_MEAN_train = 1 / (1 + np.exp(-X_train.dot(beta_post_mean)))
p_MAP_train = 1 / (1 + np.exp(-X_train.dot(map_params['beta'])))
p_CDF_train = (train_trace['p'] > 0.5).mean(axis=0)
p_PP_train = y_train_PP['Y'].mean(axis=0)
p_dict_train = {'p_MEAN': p_MEAN_train, 'p_MAP': p_MAP_train, 'p_CDF': p_CDF_train, 'p_PP': p_PP_train}
```

```python
fig, ax = plt.subplots(1, 4, figsize=(18, 4))
plt.suptitle(r'Distribution of p from Different $\beta$ Estimates (on train set)', fontsize=16, weight='heavy')
plt.subplots_adjust(top=0.8)
for i, (k, v) in enumerate(p_dict_train.items()):
    ax[i].hist(v, alpha=0.5, bins=50)
    ax[i].set_title(k, fontsize=16)
plt.show()
```

```python
# 2.9 2.10 distribution of misclassification rates
mis_pp_train = (y_train_PP['Y'] != np.tile(y_train, [10000, 1])).mean(axis=1)
mis_p_post_train = ((train_trace['p']>0.5) != np.tile(y_train, [10000, 1])).mean(axis=1)

fig, ax = plt.subplots(1, 1, figsize=(7, 5))
ax.hist(mis_pp_train, alpha=0.5, bins=15, label='mis rate = P[ pp_samples != y ]')
ax.hist(mis_p_post_train, alpha=0.5, bins=10, label='mis rate = P[ I(p_post>0.5) != y ]')
ax.set_title('Distribution of Misclassification Rates (on train set)', fontsize=15)
plt.legend(fontsize=12)
plt.tight_layout()
```

### Answer 2.9 2.10

On the **training** set, the distribution of misclassification rates from posterior predictive samples is wider than that from using $\beta$ posteriors.

```python
# posterior predictive samples on test set
X_shared.set_value(X_test.values)
y_test_PP = pm.sample_ppc(train_trace, model=iris_model, samples=10000)
y_test_PP['Y'].shape
```

```python
# X_test.dot(beta_posterior)
p_post_test = 1 / (1 + np.exp(-X_test.dot(train_trace['beta'].T).T))
p_post_test.shape
```

```python
# 2.11 distribution of misclassification rates - test set
p_MEAN_test = 1 / (1 + np.exp(-X_test.dot(beta_post_mean)))
p_MAP_test = 1 / (1 + np.exp(-X_test.dot(map_params['beta'])))
p_CDF_test = (p_post_test > 0.5).mean(axis=0)
p_PP_test = y_test_PP['Y'].mean(axis=0)
p_dict_test = {'p_MEAN': p_MEAN_test, 'p_MAP': p_MAP_test, 'p_CDF': p_CDF_test, 'p_PP': p_PP_test}

fig, ax = plt.subplots(1, 4, figsize=(18, 4))
plt.suptitle(r'Distribution of p from Different $\beta$ Estimates (on test set)', fontsize=16, weight='heavy')
plt.subplots_adjust(top=0.8)
for i, (k, v) in enumerate(p_dict_test.items()):
    ax[i].hist(v, alpha=0.5, bins=50)
    ax[i].set_title(k, fontsize=16)
plt.show()
```

3.5. Now let us think of a different problem, a regression problem in which our aim is to predict and do inference on what rating a given user of Yelp might give a particular restaurant. How might you estimate the across-user quality of a restaurant from data? And how might you estimate the across-restaurant curmudgeonlyness of a user?

**Answer 3.5**

Assuming that food and service ratings have already extracted from the reviews, we would use a hierarchical model where each user has a prior $\theta_i$ and a review rating $y_i \sim f(\theta_i)$. The procedures we propose are:

- For each user, gather all reviews written by him/her for all restaurants

- Compute the prior parameter of this user. For example, we model the mean rating of all users came from a normal distribution $\theta_i \sim \mathcal{N}(\mu_0, \sigma_0)$

- The specific review ratings of user i, ($\vec{y_{ij}}$, a 2D vector representing food and service ratings), came from another normal distribution $\vec{y_{ij}} \sim \mathcal{N}(\theta_i, \sigma_1)$

(Assuming $\sigma_0$ and $\sigma_1$ are given fixed parameters.)

3.6 Additionally, consider a "space of latent factors" where aspects of the user's taste interact with aspects of the restaurant. An example of such a factor might be the user's propensity to get emotional after having the perfect filet-mignon. How might you combine this information with that in 3.5 to improve your prediction and inference?

**Answer 3.6**

- For each user, augment the user prior parameters to 2D, ($\theta_{i0}, \theta_{i1}$), where $\theta_{i0}$ is the mean rating for a non-emotional user and $\theta_{i1}$ is the mean rating for an emotional user. ($\theta_{i0}, \theta_{i1}$) came from a joint normal distribution:

$$
\begin{pmatrix}
\theta_{i0} \\
\theta_{i1}
\end{pmatrix} \sim \mathcal{N} 
\left(
\begin{pmatrix}
\mu_0 \\
\mu_1
\end{pmatrix}, 
\Sigma_0
\right)
$$

- The specific review ratings of user i, $\vec{y_{ij}} = (\vec{y_{ij}^0}, \vec{y_{ij}^1})$, would be a 2 x 2D vector representing food and service ratings under non-emotional and emotional scale, each of which independently came from a normal distribution:

$$\vec{y_{ij}^0} \sim \mathcal{N}(\theta_{i0}, \sigma_1)$$ 

$$\vec{y_{ij}^1} \sim \mathcal{N}(\theta_{i1}, \sigma_2)$$

(Assuming $\Sigma_0$, $\sigma_1$ and $\sigma_2$ are given fixed parameters.)
